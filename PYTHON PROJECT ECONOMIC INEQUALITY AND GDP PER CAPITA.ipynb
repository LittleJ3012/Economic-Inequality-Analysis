```python
#CREATING PLOTS ON IMPORTANT INFORMATION REGARDING GLOBAL INCOME INEQUALITY STATISTICS:
```

```python
#Cleasing the data to remove empty fields of the CSV file

gini.info()
gini.isna().sum()
```

```python
#Colleting data about U.S gini index
gini_us = gini[gini['country'] == 'United States']
gini_us.head()
```

```python
#Colleting data about gdp per capita in the United States
gdp_us = gdp[gdp['Entity'] == 'United States']
gdp_us.head()
```

```python
#Colleting data about U.S gdp per capita from 1950 to 2022
gdp_us_1950_2022 = gdp_us[(gdp_us['Year'] >= 1950) & (gdp_us['Year'] <= 2022)]
gdp_us_1950_2022.head()
```

```python
#Plotting a chart about the changes on GINI index in the U.S troughout the years

plt.figure(figsize=(15, 9))
plt.plot(gini_us['year'], gini_us['p0p100_gini_pretax'], marker='o')

plt.title('United States Gini index score troughout the years')
plt.xlabel('Year')
plt.ylabel('Gini index')
plt.grid(True)

plt.show()
```

```python
#Plotting a chart about the changes on GDP per capta in the U.S troughout the years (2000-2022)

plt.figure(figsize=(15, 9))
plt.plot(gdp_us_1950_2022['Year'], gdp_us_1950_2022['GDP per capita'], marker='o')

plt.title('U.S GDP per capita troughout the years')
plt.xlabel('Year')
plt.ylabel('GDP per capita')
plt.grid(True)

plt.show()
```

```python
#Creating a chart on the ten worst (most unequal) countries on GINI index 

gini_latest = gini.sort_values('year').groupby('country').tail(1)

#The Lower the Gini Index, the more economically equal the country is 
df_top10 = gini_latest.sort_values('p0p100_gini_pretax', ascending=False).head(10); #Selects the ten worst countries

plt.figure(figsize=(15, 8))
plt.bar(df_top10['country'], df_top10['p0p100_gini_pretax'], color='red')

plt.title('The Ten Most Unequal Countries / Regions (Highest Gini Index) - 2023')
plt.xlabel('Country')
plt.ylabel('Gini Index (pretax)')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

```python
#Creating a chart on the ten best (most equal) countries on GINI index 

gini_latest = gini.sort_values('year').groupby('country').tail(1)

#The Lower the Gini Index, the more economically equal the country is 
df_top10 = gini_latest.sort_values('p0p100_gini_pretax', ascending=True).head(10); #Selects the ten best countries

plt.figure(figsize=(15, 8))
plt.bar(df_top10['country'], df_top10['p0p100_gini_pretax'], color='green')

plt.title('The Ten Most Equal Countries (Lowest Gini Index) -- 2023')
plt.xlabel('Country')
plt.ylabel('Gini Index (pretax)')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

```python
#Creating a chart on the ten worst countries in the GDP per capita rank (lowest GDP per capita) in 2022

gdp_latest = gdp.sort_values('Year').groupby('Entity').tail(1)

df_top10 = gdp_latest.sort_values('GDP per capita', ascending=True).head(10); #Selects the ten worst countries
plt.figure(figsize=(15, 8))
plt.bar(df_top10['Entity'], df_top10['GDP per capita'], color='red')

plt.title('The Ten Countries / Regions with the Lowest GDP Per Capita in The World - 2022')
plt.xlabel('Country')
plt.ylabel('GDP per capita value (in US$ dollars)')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

```python
#Creating a chart on the ten best countries in the GDP per capita rank (highest GDP per capita) in 2022

gdp_latest = gdp.sort_values('Year').groupby('Entity').tail(1)

df_top10 = gdp_latest.sort_values('GDP per capita', ascending=False).head(10); #Selects the ten best countries
plt.figure(figsize=(15, 8))
plt.bar(df_top10['Entity'], df_top10['GDP per capita'], color='green')

plt.title('The Ten Countries / Regions with the Highest GDP Per Capita in The World - 2022')
plt.xlabel('Country')
plt.ylabel('GDP per capita value (in US$ dollars)')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```
